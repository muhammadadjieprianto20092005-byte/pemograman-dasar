```python
import pandas as pd

city_names = ['Jakarta', 'Bandung', 'Makassar', 'Surabaya', 'Medan', 'Yogyakarta', 'Malang']
population = [498044, 964254, 491918, 8398748, 653115, 883305, 744955]
num_airports = [2, 2, 8, 3, 1, 3, 2]

df = pd.DataFrame({
    'City Name': city_names,
    'Population': population,
    'Airports': num_airports,
})

df
```

```python
# Ringkasan statistik
df.describe()

# Rata-rata populasi
rata_rata_populasi = df['Population'].mean()
print("Rata-rata populasi:", rata_rata_populasi)

# Modus jumlah bandara
modus_airports = df['Airports'].mode()
print("Modus Airports:", modus_airports)
```

```python
1. Kota mana yang populasinya mendekati rata-rata?
=populasi terdekat dengan rata-rata adalah Bandung

2. Apa arti modus untuk kolom Airports?
=angka yang paling sering muncul pada kolom airports
```

```python
import matplotlib.pyplot as plt

df['Population'].hist()
plt.title("Distribusi Populasi Kota")
plt.xlabel("Jumlah Penduduk")
plt.ylabel("Frekuensi")
plt.show()
```

Tuliskan pengamatan dari histogram tersebut
= Distribusi tidak merata ada satu kota dengan populasi sangat besar yaitu Surabaya, sehingga histogram akan skewed miring ke kanan,Kota lain berada di kelompok populasi kecil menengah dan mayoritas kota hanya memiliki populasi kurang dari 1 juta.

```python
# Korelasi
corr_matrix = df[['Population', 'Airports']].corr()
print("Korelasi:", corr_matrix)

# Kovarian
cov_matrix = df[['Population', 'Airports']].cov()
print("Kovarian:", cov_matrix)
```

1. Apakah korelasinya positif atau negatif?
=Karena kota yang besar biasanya punya bandara lebih banyak dan korelasinya positif
2. Apakah kota dengan penduduk besar cenderung punya lebih banyak bandara?
=Iya,Kota dengan populasi besar seperti Surabaya memiliki jumlah bandara lebih banyak dibandingkan kota kecil

```python
# Menambah kolom kategori
df['Kategori_Bandara'] = df['Airports'].apply(lambda x: 'Banyak' if x >= 3 else 'Sedikit')
df

# Rata-rata populasi per kategori
group_mean = df.groupby('Kategori_Bandara')['Population'].mean()
print(group_mean)

# Agregasi lengkap
group_agg = df.groupby('Kategori_Bandara').agg({
    'Population': ['mean', 'max'],
    'Airports': ['mean', 'count']
})
group_agg
```

1. Bandingkan populasi kota dengan kategori Banyak dan Sedikit.
=- Banyak lebih dari 3 airport >	Surabaya, Bandung	Populasi rata-rata tinggi
- sedikit kurang dari 3 airport < kota lain Populasi lebih rendah

2. Apa kesimpulannya?
=Semakin besar kota maka semakin besar kebutuhan bandara dan jumlah bandara meningkat
