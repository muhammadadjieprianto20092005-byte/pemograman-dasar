```python
# INPUT SUHU
suhu = float(input("input suhu air dingin (°C): "))
if suhu <= 90:
    print("STATUS: Normal")
elif suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")
```

```python
# Data hasil vibrasi dari 5 unit mesin
vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

for i in range(len(vibrasi_mesin)):
    if vibrasi_mesin[i] > 2.0:
        print(f"Warning: Unit {i+1} vibrasi tinggi!")
```

```python
# Suhu awal
suhu_saat_ini = 60

# Naikkan suhu sebesar 5°C tiap iterasi

while suhu_saat_ini < 85:
    print(f"Suhu saat ini: {suhu_saat_ini}°C")
    suhu_saat_ini += 5

print("Target suhu 85°C telah tercapai.")
```

```python
# input masukkan
kapasitas_maks = float(input("Masukkan kapasitas maksimum (kW): "))
beban_terpakai = float(input("Masukkan beban terpakai (kW): "))

# persentas pemakaian
persentase = (beban_terpakai / kapasitas_maks) * 100

print(f"Persentase pemakaian: {persentase:.2f}%")

# Percabangan status beban
if persentase <= 60:
    print("Status: Beban Ringan")
elif persentase <= 85:
    print("Status: Beban Sedang")
else:
    print("Status: Beban Berat")
```

```python
# Daftar suhu di 5 titik boiler
suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]

# Looping untuk menampilkan setiap suhu
for i in range(len(suhu_boiler)):
    print(f"Titik {i+1}: {suhu_boiler[i]}°C")

    # Beri peringatan jika suhu melebihi 100°C
    if suhu_boiler[i] > 100:
        print("Warning: Suhu melebihi 100°C!\n")
    else:
        print("Status: Normal\n")
```
