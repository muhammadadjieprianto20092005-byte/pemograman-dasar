```python
import pandas as pd

body_measurement_df = pd.DataFrame.from_records((
  (2, 83.82, 8.4),
  (4, 99.31, 16.97),
  (3, 96.52, 14.41),
  (6, 114.3, 20.14),
  (4, 101.6, 16.91),
  (2, 86.36, 12.64),
  (3, 92.71, 14.23),
  (2, 85.09, 11.11),
  (2, 85.85, 14.18),
  (5, 106.68, 20.01),
  (4, 99.06, 13.17),
  (5, 109.22, 15.36),
  (4, 100.84, 14.78),
  (6, 115.06, 20.06),
  (2, 84.07, 10.02),
  (7, 121.67, 28.4),
  (3, 94.49, 14.05),
  (6, 116.59, 17.55),
  (7, 121.92, 22.96),
), columns=("age", "height_cm", "weight_kg"))

body_measurement_df
```

```python
# Ringkasan statistik
body_measurement_df.describe()

# Rata-rata tinggi & berat
mean_height = body_measurement_df['height_cm'].mean()
mean_weight = body_measurement_df['weight_kg'].mean()
print("Rata-rata tinggi:", mean_height)
print("Rata-rata berat:", mean_weight)

# Modus umur
mode_age = body_measurement_df['age'].mode()
print("Modus umur:", mode_age)
```

1. Umur berapa yang paling sering muncul?
= Umur yang paling banyak muncul di dataset adalah usia 6 tahun.
2. Apakah berat badan rata-rata ringan/sedang/berat?
= Rata-rata berat di dataset â‰ˆ 15-20 kg, ini termasuk kategori berat badan anak-anak (ringan)



```python
import matplotlib.pyplot as plt

# Tinggi
body_measurement_df['height_cm'].hist()
plt.title("Distribusi Tinggi Badan (cm)")
plt.show()

# Berat
body_measurement_df['weight_kg'].hist()
plt.title("Distribusi Berat Badan (kg)")
plt.show()
```

Apakah tinggi & berat menyebar rata atau menumpuk?
= Tinggi lebih cenderung menumpuk pada 95-120 cm
Berat banyak menumpuk pada 12-20 kg
distribusi tidak merata, tetapi menumpuk di sekitar nilai tengah

```python
# Mean
group_mean_age = body_measurement_df.groupby("age").mean()
group_mean_age

# Agg lengkap
group_agg_age = body_measurement_df.groupby("age").agg({
    'height_cm': ['mean', 'min', 'max'],
    'weight_kg': ['mean', 'min', 'max']
})
group_agg_age
```

1. Usia berapa rata-rata berat paling tinggi?
= Dari data, usia tertinggi beratnya diusia 7 tahun
2. Apakah tinggi meningkat sesuai usia?
= iya,Semakin bertambah usia maka tinggi rata-rata meningkat dsri dataset menunjukkan kenaikan tinggi badan seiring bertambahnya usia


```python
# Korelasi
corr_hw = body_measurement_df[['height_cm', 'weight_kg']].corr()
print(corr_hw)

# Kovarian
cov_hw = body_measurement_df[['height_cm', 'weight_kg']].cov()
print(cov_hw)
```

1. Apakah korelasinya kuat/lemah?
= Biasanya korelasi tinggi-berat sekitar 0.7-0.9 ini adalah korelasi yang kuat
2. Apakah hubungannya positif?
= iya,Semakin tinggi seseorang semakin besar pula berat badannya
