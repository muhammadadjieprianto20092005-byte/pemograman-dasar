```python
print("=" * 30)
print("SELAMAT DATANG DI KAFE GEMINI")
print("=" * 30)
print("Menu Makanan & Minuman")
print("1. Kopi Hitam - Rp 12.000")
print("2. Teh Manis - Rp 8.000")
print("3. Roti Bakar - Rp 15.000")
print("4. Mie Instan - Rp 10.000")
print("=" * 30)

pilihan_str = input("Silakan masukkan nomor pilihan Anda (1-4): ")
pilihan = int(pilihan_str)
nama_pesanan = ""
harga = 0

if pilihan == 1:
    nama_pesanan = "Kopi Hitam"
    harga = 12000
    print(f"Anda memesan: {nama_pesanan} dengan harga Rp {harga}")
elif pilihan == 2:
    nama_pesanan = "Teh Manis"
    harga = 8000
    print(f"Anda memesan: {nama_pesanan} dengan harga Rp {harga}")
elif pilihan == 3:
    nama_pesanan = "Roti Bakar"
    harga = 15000
    print(f"Anda memesan: {nama_pesanan} dengan harga Rp {harga}")
elif pilihan == 4:
    nama_pesanan = "Mie Instan"
    harga = 10000
    print(f"Anda memesan: {nama_pesanan} dengan harga Rp {harga}")
else:
    print("Maaf, pilihan menu tidak valid. Silakan jalankan ulang program.")

if harga > 0:
    print("=" * 30)

    # Input jumlah pesanan
    jumlah_str = input("Masukkan jumlah pesanan: ")
    jumlah = int(jumlah_str)

    # Hitung total harga sebelum diskon
    total_sebelum_diskon = harga * jumlah
    print(f"Total sebelum diskon: Rp {total_sebelum_diskon}")

    # SISTEM DISKON 3 KATEGORI - TAMBAHAN BARU
    if total_sebelum_diskon >= 100000:
        diskon = 0.70  # 70%
        kategori_diskon = "Diskon Platinum 70%"
    elif total_sebelum_diskon >= 50000:
        diskon = 0.50  # 50%
        kategori_diskon = "Diskon Gold 50%"
    elif total_sebelum_diskon >= 30000:
        diskon = 0.30  # 30%
        kategori_diskon = "Diskon Silver 30%"
    else:
        diskon = 0
        kategori_diskon = "Tidak ada diskon"

    # Hitung diskon dan total setelah diskon
    besar_diskon = total_sebelum_diskon * diskon
    total_setelah_diskon = total_sebelum_diskon - besar_diskon

    print(f"Kategori diskon: {kategori_diskon}")
    print(f"Besarnya diskon: Rp {besar_diskon:,.0f}")
    print(f"Total yang harus dibayar: Rp {total_setelah_diskon:,.0f}")

    print("=" * 30)
    uang_dibayar_str = input(f"Masukkan jumlah uang Anda: Rp ")
    uang_dibayar = int(uang_dibayar_str)

    if uang_dibayar >= total_setelah_diskon:
        kembalian = uang_dibayar - total_setelah_diskon
        print("\n=== STRUK PEMBAYARAN ===")
        print(f"Pesanan Anda     : {nama_pesanan}")
        print(f"Jumlah           : {jumlah} porsi")
        print(f"Harga satuan     : Rp {harga}")
        print(f"Total sebelum diskon: Rp {total_sebelum_diskon}")
        print(f"Kategori diskon  : {kategori_diskon}")
        print(f"Besarnya diskon  : Rp {besar_diskon:,.0f}")
        print(f"Total harga      : Rp {total_setelah_diskon:,.0f}")
        print(f"Uang Pembayaran  : Rp {uang_dibayar}")
        print(f"Kembalian        : Rp {kembalian}")
        print("=" * 30)
        print("Terima kasih telah berbelanja!")
    else:
        kekurangan = total_setelah_diskon - uang_dibayar
        print(f"Maaf, uang Anda kurang sebesar Rp {kekurangan}.")

print("=" * 80)
```
